```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('/content/drive/MyDrive/customer/customercon_dataset2_6.xlsx')

# Set plot size
plt.figure(figsize=(14, 8))

# Create a pairplot where target column is colored
pairplot = sns.pairplot(data, x_vars=['aitechnology', 'blockchaintechnology', 'cloudcomputingtech', 'bigdatatechnology', 'digitaltechapplication'],
                        y_vars='customerconcentration',
                        height=5, aspect=0.8, kind='scatter')

# Add a title to the entire plot
plt.suptitle('Customer Concentration vs Digital Transformation Technologies', y=1.02)
plt.tight_layout()

# Save the figure as a PDF file with 400 DPI
pairplot.savefig('/content/drive/MyDrive/customer/CustomerConcentration_vs_Technologies.pdf', format='pdf', dpi=500)

# Display the plot
plt.show()
```
